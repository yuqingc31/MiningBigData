# Big Data Visualisation

In today's data-driven world, the sheer volume, variety, and velocity of data generated from various sources have given rise to the concept of "big data." Big data refers to vast datasets that are too large and complex to be effectively analyzed using traditional data processing and visualization methods. To make sense of this wealth of information, organizations and analysts turn to big data visualizationâ€”a powerful tool that transforms massive datasets into meaningful insights, patterns, and trends through visual representations.

Big data visualization is a critical component of the data analytics process, serving as a bridge between the raw data and actionable insights. By employing advanced visualization techniques and tools designed to handle large-scale data, it enables users to explore, analyze, and interpret complex information efficiently.

### Key Components of Big Data Visualization:

**Large Datasets:** Big data visualization deals with datasets of immense size, often containing millions, billions, or even trillions of records. These datasets can come from various sources, such as social media, sensors, IoT devices, and transaction logs.

**Data Processing:** Managing and processing such massive datasets require specialized technologies like distributed computing frameworks (e.g., Apache Hadoop, Spark) and cloud-based solutions. Data preprocessing, cleaning, and transformation are essential steps to prepare the data for visualization.

**Visualization Tools:** Big data visualization tools and platforms are specifically designed to handle the scale and complexity of big data. They offer a wide range of visualization options, including interactive dashboards, charts, graphs, heatmaps, and geospatial visualizations.

**Interactivity:** Interactivity is a hallmark of big data visualization. Users can drill down into data subsets, apply filters, and explore different dimensions of the dataset in real-time. This feature empowers analysts to gain deeper insights and make data-driven decisions.

**Complex Visual Representations:** Big data visualization often involves complex visual representations to convey multifaceted information effectively. This may include network graphs, 3D visualizations, time-series analyses, and data-driven storytelling.

### Benefits of Big Data Visualization:

**Data Exploration:** Big data visualization allows users to explore vast datasets quickly, identifying patterns, outliers, and correlations that might be hidden in the raw data.

**Decision Support:** It provides decision-makers with clear, visual insights into complex data, enabling them to make informed decisions faster.

**Real-time Insights:** Many big data visualization tools offer real-time or near-real-time updates, allowing organizations to respond rapidly to changing conditions.

**Improved Communication:** Visualizations simplify the communication of data-driven findings across teams, departments, and stakeholders, making it accessible to a broader audience.

**Predictive Analytics:** Visualization is a vital component of predictive analytics, helping organizations forecast trends, detect anomalies, and plan for the future.

In conclusion, big data visualization is a crucial technique for harnessing the potential of big data. It empowers organizations to unlock valuable insights, optimize operations, and drive innovation in an increasingly data-centric world. By converting vast and complex datasets into clear and actionable visuals, big data visualization plays a pivotal role in informed decision-making and business success.

### Let's examine visualisation of random data to learn the techniques and code

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data
np.random.seed(0)
data_size = 1000
random_data = {
    'Age': np.random.randint(18, 65, data_size),
    'Income': np.random.randint(20000, 100000, data_size),
    'Spending': np.random.randint(500, 10000, data_size)
}

# Create a DataFrame
df = pd.DataFrame(random_data)

df["Income"] = df["Income"]*df["Age"]/40
df["Spending"] = df["Spending"]*(1-((df["Age"]-40)/25)**2)*(1+df["Income"]/max(df["Income"]))


# Workshop Part 1: Data Exploration
# - Display basic statistics about the dataset
print(df.describe())

# Workshop Part 2: Data Visualization
# - Create visualizations to explore the data
# - Use Matplotlib and Seaborn for plotting

# my work: 

plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Spending', data=df)
plt.title('my work spending Distribution by age')
plt.xlabel('Age')
plt.ylabel('Spending')
plt.xticks(rotation=90)
plt.show()


# Example 1: Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Example 2: Pairplot for Age, Income, Score, and Spending
sns.pairplot(df[['Age', 'Income', 'Spending']])
plt.suptitle('Pairplot of Age, Income, and Spending', y=1.02)
plt.show()

# Example 3: Boxplot of Income by Age Group
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Income', data=df)
plt.title('Income Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Income')
plt.xticks(rotation=90)
plt.show()

# Workshop Part 3: Interpretation and Discussion
# - Discuss the insights gained from the visualizations
# - Explore relationships and patterns in the data

# my answers:
# Consumers are mainly young people. Older persons and infants consume less


# Example 4: Correlation Heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

```

**t-Distributed Stochastic Neighbour Embedding (t-SNE)**
Like PCA, t-SNE is another dimensionality reduction technique, developed in part by the godfather of machine learning himself, Geoffrey Hinton. Quoting Wikipedia : It is a nonlinear dimensionality reduction technique well-suited for embedding high-dimensional data for visualization in a low-dimensional space of two or three dimensions. Specifically, it models each high-dimensional object by a two- or three-dimensional point in such a way that similar objects are modeled by nearby points and dissimilar objects are modeled by distant points with high probability.

Traditionally t-SNE is applied on data which has already been reduced dimensionally by other means such as PCA. This is what we will do ! We will apply 2-component t-SNE on the outputs of the 20-component PCA. This will allow us to plot the t-SNE representation in 2D space.

Consider what are the applications of such a technique? Where would you recommend this technqiue?


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Step 1: Apply PCA for dimensionality reduction to 2 components
pca = PCA(n_components=2)  
X_pca = pca.fit_transform(X) 

# Step 2: Apply t-SNE for further visualization
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_pca)

# Visualize the results
plt.figure(figsize=(12, 6))

# Plot PCA results
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title('PCA Visualization')

# Plot t-SNE results
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
plt.title('t-SNE Visualization')

plt.show()
```

```python
import pandas as pd

# To ignore warnings, use the following code to make the display more attractive.
# Import seaborn and matplotlib.
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)

#To import the Iris dataset:
iris = pd.read_csv("Iris.csv") # the iris dataset is now a Pandas DataFrame
```

In Pandas use Andrews Curves to plot and visualize data structure.
Each multivariate observation is transformed into a curve and represents the coefficients of a Fourier series.
This useful for detecting outliers in times series data.
Use colormap to change the color of the curves

```python

from pandas.plotting import andrews_curves
andrews_curves(iris.drop("Id", axis=1), "Species",colormap='rainbow')
plt.show()
```

Parallel_cordinates plot each feature on a seperate column.

Each feature is then connected by lines, for each data sample

Again, colormap can be used to choose an assortment of colors. 

```python
from pandas.plotting import parallel_coordinates
parallel_coordinates(iris.drop("Id", axis=1), "Species",colormap='cool')
plt.show()
```
